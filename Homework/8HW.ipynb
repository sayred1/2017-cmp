## 8.5 Homework
Write your own program to do the followings
- generate a data with random noise
- choose different polynomial function to fit the data
- discuss which polynomial function yields the best fit and give your reasoning.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

Min = -4.0
Max = 4.0
N = 100
order = [0,1,2,3,4,5,6,7,8]

#function
f = lambda x: 0.1*np.sin(x) * x**2

#noisy data
x = np.linspace(Min,Max,N)
y = f(x) + 0.1*np.random.randn(N) - 0.01

#lists of all fits and their y values 
fits = []
y_ps = []

#use of the polyfit library to fit the function to 11 different polynomial functions
for o in order:
    fits.append(np.polyfit(x, y, o, full=True))
    y_ps.append(np.polyval(fits[o][0], x))
# plot polyfits
fig, axs = plt.subplots(3, 3, figsize=(20, 20))
plt.title('0.1*np.sin(x) * x**2')
count = 0
for i in range(len(axs)):
    for j in range(len(axs[i])):
        axs[i,j].set_title('polyfit order %s, resuduals %s' %(str(count), fits[count][1]))
        axs[i,j].plot(x,y,'o',color = 'k', label = 'noisy data', markersize = 1)
        axs[i,j].plot(x,f(x), label='ideal function')
        axs[i,j].plot(x,y_ps[count], label='polyfit'+str(i))
        axs[i,j].legend(fontsize=12)
        count += 1
fig.suptitle('f(x) = 0.1 * x * x * sin(x) - noise')
plt.show()
```

The polyfit function that seems to accurately fit the data around the 6th order. Beyond may be overfitting.
