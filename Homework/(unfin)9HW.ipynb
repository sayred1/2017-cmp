```python
%matplotlib inline  
import matplotlib.pyplot as plt
import numpy as np
from numpy import fft

data = loadtxt("circular.txt",float)
imshow(data)
plt.show()

c = fft.ifft(data)
x = np.linspace(0,1000,np.shape(data)[0])
plt.figure(figsize=(15,5))
plt.semilogx(x, data, 'k--')
plt.show()
plt.figure(figsize=(15,5))
plt.semilogx(x,abs(c),'k--')
plt.show()
```
