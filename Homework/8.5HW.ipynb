Write your own program to do the followings
- generate a data with random noise
- choose different polynomial function to fit the data
- discuss which polynomial function yields the best fit and give your reasoning.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

Min = -4.0
Max = 4.0
N = 100
order = []

#specified orders from 0 - 10 to define different polynomial functions
for i in range(0,11):
    order.append(i)

#function
f = lambda x: x**2

#noisy data
x = np.linspace(Min,Max,N)
y = f(x) + np.random.rand(N) - 0.5

#lists of all fits and their y values 
fits = []
y_ps = []

#use of the polyfit library to fit the function to 11 different polynomial functions
for o in order:
    fits.append(np.polyfit(x, y, o, full=True))
    y_ps.append(np.polyval(fits[o][0], x))

#for loop to plot each polynomial fit of order 0 - 10 along with the noisy data and the ideal functions
for i in range(len(y_ps)):
    plt.figure(figsize = (20,10))
    plt.title('fitted coefficients' + str(fits[i][0]) + '\n residual: ' + str(fits[i][1]))
    plt.plot(x,y,'o',color = 'k', label = 'noisy data')
    plt.plot(x,f(x), label='ideal function')
    plt.plot(x,y_ps[i], label='polyfit'+str(i))
    plt.legend(fontsize=12)
    plt.show()
    
#below are 11 figures of the same noisy data and expected function but different polynomial fittings,
#which differ in order. As can be seen, as the order increases, the accuracy of the fitting increases.
#This is seen in the plots as the orange line fitting within the data more and more as the order
#increases. Furthermore, the at an order of 10, the fitting gives a better residual amount in 
#comparison to the noisy data, which is a measure of agreement between the two. Thus, compared
#to the noisy data, the polynomail with order 10 gives the best fitting. 

```
