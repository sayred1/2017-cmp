# Gradient Descent

Optimization method that looks at the negative gradient (or downward gradient) at the current point on a functionâ€™s curve.

The algorithm iteratively updates the point until the coordinates obtained are the function's minimum.

For a differentiable function relying on a single variable, the root we are looking for is given by 

$$x_{n+1} = x_n - \alpha_n \Delta_x F(x_n)$$

Where $x_n$ is the point previously evaluated, $\alpha_n$ is the step size, $\Delta_x = \frac{d}{dx}$, and $x_{n+1}$ is the updated point. This can be generalized to a multi-variable function with a similar update of each variable.

The hyperparameter $\alpha$ may be static throughout the optimization process or may change iteratively by the following 

$$\alpha_n = \frac{|(x_n-x_{n-1})^T(\Delta_x F(x_n) - \Delta_xF(x_{n-1})|}{||  \Delta_x F(x_n) - \Delta_xF(x_{n-1})   ||^2}$$

### Example of gradient descent in 1-D for a simple function.
    1. point starts at some initial x_0
    2. The function and its derivative are evaluated at that point, and until convergence, a new point is generated based on the above formula.
$\alpha$ is iteratively updated.

# Linear Conjugate Gradient Descent

The Conjugate Gradient Method is the most prominent iterative method for solving sparse systems of linear equations.

Effective for systems of the form $Ax=b$ where $x$ is an unknown vector, $b$ is a known vector, and $A$ is a known, square, symmetric, positive-definite matrix. 

Recall, from linear algebra, that $Ax=b$ written out fully is 

$$\left[\begin{array}
{rrr}
A_{11} & A_{12} & ... & A_{1n} \\
A_{21} & A_{22} &     & A_{2n} \\
\vdots &        & \ddots & \vdots \\
A_{n1} & A_{n2} & ... & A_{nn}
\end{array}\right]
\left[\begin{array}
{rrr}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{array}\right] 
=
\left[\begin{array}
{rrr}
b_1 \\
b_2 \\
\vdots \\
b_n
\end{array}\right]
$$

Limitations to the linear conjugate gradient descent arise from the requirement of positive-definiteness of A,

$$x^TAx > 0$$
